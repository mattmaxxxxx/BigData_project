```
from pyspark.sql.functions import col, translate, percent_rank, unix_timestamp
from pyspark.sql.types import StructType, StructField, StringType,  DateType, FloatType, TimestampType, IntegerType
from pyspark.sql.window import Window

schema = StructType([
    StructField("Date",TimestampType()),
    StructField("High",FloatType()),
    StructField("Low",FloatType()),
    StructField("Open",FloatType()),
    StructField("Close",FloatType()),
    StructField("Volume",IntegerType()),
    StructField("Adj Close",FloatType()),
    StructField("company_name",StringType())
])

filePath = 'dbfs:/FileStore/shared_uploads/matthieu.schlienger@epita.fr/stocks_data/GOOGLE.csv'

df = spark.read.csv(filePath, header=True, schema=schema)

df = df.withColumn('rank', percent_rank().over(Window.partitionBy().orderBy('Date')))
df = df.withColumn('DateInt', unix_timestamp('Date', format='yyyy-MM-dd HH:mm:ss.000'))# Convert the date to int for the purpuse of the regression

display(df)
```

```
trainDF = df.where('rank <= 0.8').drop('rank')
testDF = df.where('rank > 0.8').drop('rank')
print(f"There are {trainDF.cache().count()} rows in the training set, and {testDF.cache().count()} in the test set")
```

```
display(trainDF.select("Open", "Close").summary())
```

```
display(testDF.select("Open", "Close").summary())
```

```
from pyspark.ml.feature import VectorAssembler

vecAssembler = VectorAssembler(inputCols=['DateInt', 'Open'], outputCol='features')
vecTrainDF = vecAssembler.transform(trainDF)
vecTrainDF.select('DateInt', 'Open', 'features', 'Close').show(10)

vecTrainDF.select("Open", "features", "Volume").show(10)
```

```
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline

lr = LinearRegression(featuresCol='features', labelCol='Close', predictionCol='Close prediction')

pipeline = Pipeline(stages=[vecAssembler, lr])
pipelineModel = pipeline.fit(trainDF)

predDF = pipelineModel.transform(testDF)
```

```
from pyspark.ml.evaluation import RegressionEvaluator

regressionEvaluator = RegressionEvaluator(predictionCol="Close prediction",
                                          labelCol="Close",
                                          metricName="rmse")
rmse = regressionEvaluator.evaluate(predDF)

r2 = RegressionEvaluator(predictionCol="Close prediction",
                         labelCol="Close",
                         metricName="r2").evaluate(predDF)

print(f"RMSE is {rmse:.1f}")
print(f"R2 is {r2}")
```

```
import matplotlib.pyplot as plt

y = [val for val in df.select('Close').collect()]
x_ts = [val for val in df.select('Date').collect()]

y_pred = [val for val in predDF.select('Close prediction').collect()]

plt.plot(x_ts[-198:], y[-198:],'b', label = "actual close price")
plt.plot(x_ts[-198:],y_pred, 'r', label = "predicted close price")

plt.ylabel('Close')
plt.xlabel('timestamp')
plt.title('Close value of Google')
plt.legend(loc='upper left')

plt.show()
```
